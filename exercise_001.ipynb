```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///taskeasy.db', echo=False)
```

# Question 1
A query and result to show every product/coupon code combination with the resulting cost after the discount is applied

```python
query_q1 = """
    SELECT
        A.product_name,
        A.cost AS original_cost,
        B.coupon_code,
        B.discount_percentage * 100 AS discount_percentage,
        A.cost * (1 - B.discount_percentage) AS cost_discount_applied
    FROM
        Products A,
        Discounts B
"""
```

```python
q1 = pd.read_sql(query_q1, con=engine)
q1.head(n=15)
```

```python
len(q1[q1.product_name == 'excitement'])
```

# Question 2
A query and result to show the same set of data above, but only for products that would now cost less than or equal to $50.

```python
query_q2 = """
SELECT
        A.product_name,
        A.cost AS original_cost,
        B.coupon_code,
        B.discount_percentage * 100 AS discount_percentage,
        A.cost * (1 - B.discount_percentage) AS cost_discount_applied
    FROM
        Products A,
        Discounts B
    WHERE
        A.cost * (1 - B.discount_percentage) <= 50
"""
```

```python
q2 = pd.read_sql(query_q2, con=engine)
q2.head(n=15)
```
